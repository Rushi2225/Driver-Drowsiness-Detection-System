```python
import os
import glob
import shutil
import random
from tqdm import tqdm
```

```python
raw_data = r"F:\Projects\DDDS\mrlEyes_2018_01"
for dirpath, dirname, filename in os.walk(raw_data):
    for file in tqdm([f for f in filename if f.endswith('.png')]):
        if file.split('_')[4] == '0':
            path=r"F:\Projects\DDDS\Prepared Data\Close Eyes"
            if not os.path.exists(path):
                os.makedirs(path)
            shutil.copy(src=dirpath + '/' + file, dst= path)
        elif file.split('_')[4] == '1':
            path=r"F:\Projects\DDDS\Prepared Data\Open Eyes"
            if not os.path.exists(path):
                os.makedirs(path)
            shutil.copy(src=dirpath + '/' + file, dst= path)        
```
